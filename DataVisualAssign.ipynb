# Importing necessary packages and datasets

```python
import pandas as pd
#import matplotlib.pyplot as plt
import warnings 
warnings.filterwarnings("ignore")
% matplotlib inline
```

```python
# Importing dataset files
suicide = pd.read_csv("./SuicideRate.csv")
unemp = pd.read_csv("./UnempRate.csv")
```

<font size="7"> _____Data Exploring and Cleaning__________</font>

#  Suicide.csv Dataset

```python
suicide.head()
```

```python
suicide.info()
```

```python
#Explore missing data
suicide.isnull().sum()
```

```python
#Dropping missing data columns and unnecessary columns
suicide.drop(['HDI for year',' gdp_for_year ($) ','gdp_per_capita ($)', 'generation',
            'population','suicides_no', 'country-year'], axis=1, inplace=True)
suicide.rename(columns={ 'country':'Country', 'year':'Year','suicides/100k pop':'suicide_Rate'},
             inplace=True)
suicide.head()
```

```python
# Getting the column's names where age is 15-24
updatedate = suicide[(suicide['age'] != '15-24 years')].index
# Delete the other row indexes from dataFrame
suicide.drop(updatedate , inplace=True)
```

```python
# Getting the column's names where year is less than 2010
updatedate = suicide[(suicide['Year'] < 2010) ].index
# Delete these row indexes from dataFrame
suicide.drop(updatedate , inplace=True)
```

```python
# Getting the column's names where year is equal to 2015
updatedate = suicide[(suicide['Year'] == 2015) ].index 
# Delete these row indexes from dataFrame
suicide.drop(updatedate , inplace=True)
```

```python
# Getting the column's names where year is equal to  2016
updatedate = suicide[(suicide['Year'] == 2016) ].index 
# Delete these row indexes from dataFrame
suicide.drop(updatedate , inplace=True)
```

```python
#Final Prepared Suicide dataset
suicide.head()
```

-----------------------------------------------------------------------------------------------------------------

# Unemployment.csv Dataset

```python
#Viewing the Unemployment dataset
unemp.head()
```

```python
unemp.info()
```

```python
#Explore missing data
unemp.isnull().sum()
```

```python
#Dropping unnecessary columns and renaming dataframes
to_drop = ['Country Code']
unemp.rename(columns={ 'Country Name':'Country'}, inplace=True)
unemp.drop(to_drop, inplace=True, axis=1)
```

```python
# Reshapping the dataset using Pandas.DataFraem.melt
undata = pd.melt(unemp, id_vars=["Country"], value_vars=['2010', '2011', '2012', '2013',
                                                         '2014'],
                 var_name= 'Year', value_name="Unemployment_Rate")
```

```python
#Viewing the dataset after updating it
undata.head()
```

```python
#sorting the data by country and year using Pandas.DataFraem.sort_values
data_sorted = undata.sort_values(["Country", "Year"])
data_sorted.head()
```

```python
data_sorted.info()
```

```python
# Changing the Year dataFrame type from object to integer to combine it with the other dataset
data_sorted['Year'] = data_sorted['Year'].astype(int)
```

```python
#Data types after changing the Year type
data_sorted.info()
```

--------------

# Prepare Suicide.csv Dataset to Merge

```python
#group the data by country and year to sum the suicide rate using Pandas.DataFraem.groupby
suicide.groupby(['Country','Year'], as_index=False).agg({"suicide_Rate": "sum"},inplace=True)
suicide.head()
```

----------

<font size="7"> _____Merging Datasets__________</font>

```python
# Merging the two datasets to find the relationship of features using Pandas.DataFraem.merge
mergedata = pd.merge(suicide, data_sorted, on=['Country','Year'], how='outer' )
```

```python
#View the merged data
mergedata.head()
```

```python
#Checking on null valuse
mergedata.isnull().sum()
```

---------

# Cleaning Merged Datasets' Null Values

```python
# Cleaning the data from null valuse using Pandas.DataFraem.dropna
cleandata = mergedata.dropna()
```

```python
cleandata.info()
```

```python
#Check if there is any null values
cleandata.isnull().sum()
```

```python
#Final merged dataset values
cleandata.head()
```

-------

<font size="7"> _____Data Visualization__________</font>

# Suicide Rate by Gender        

```python
#Grouping the data by years and gender using pandas dataframe.groupby
gender = suicide.groupby(['Year','sex'], as_index=False).agg({"suicide_Rate": "sum"},
                                                           inplace=True)
gender
```

```python
# Reshape the dataframes to plot using pandas.DataFrame.pivot
suigendata = gender.pivot(index='Year',columns='sex', values='suicide_Rate')
```

```python
suigendata
```

```python
# Plotting the Suicide rate by gender over the years with plotly.graph

import plotly
import plotly.offline as pyo
import plotly.graph_objs as go
plotly.offline.init_notebook_mode(connected=True)



trace0 = go.Bar(
        x= suigendata.index,
        y= suigendata.male,
        name= "Male",
        marker = dict(color = 'royalblue')
)

trace1 = go.Bar(
        x= suigendata.index,
        y= suigendata.female,
        name = 'Female',
        marker = dict(color = 'lightpink')
)

data = [trace0, trace1]


layout = go.Layout(
    barmode='group',
    title='Suicide Rate by Gender<br> from 2010 to 2014',
    titlefont = dict(
         family='Old Standard TT, serif',
        size = 25,
        color = 'Black'
    ),
    xaxis=go.layout.XAxis(
        color = 'Black',
         title = "Years",
         titlefont=dict(
            family='Old Standard TT, serif',
            size=18,
            color='black'
                ),
    ),
     yaxis= dict(
         color = 'black',
         title = "Suicide Rate",
          titlefont=dict(
            family='Old Standard TT, serif',
            size=18,
            color='black'
                ),
    ),
                
)


pyo.iplot({"data":data, "layout":layout})
```

<font size="3">It is presented from the above bar chart that the majority of suicides were males compared to females during all five years from 2010 to 2014. Moreover, it is clear that the suicide rate is decreasing gradually.</font>

# Suicide and Unemployment Rate from 2010 to 2014 

```python
#Groupping the data by years to visualize the sum of rates by years
Rate_year = cleandata.groupby('Year', as_index=False).agg({"Unemployment_Rate": "sum",
                                                        "suicide_Rate": "sum"},inplace=True)
```

```python
Rate_year.head()
```

```python
# Plot rates by years with plotly.graph_objs.Scatter

trace0 = go.Scatter(
    x = Rate_year.Year,
    y = Rate_year.Unemployment_Rate,
    mode = 'markers+lines',
    name = 'Unemployment Rate'
)
trace1 = go.Scatter(
    x = Rate_year.Year,
    y = Rate_year.suicide_Rate,
    mode = 'markers+lines',
    name = 'Suicide Rate', 
)


data = [trace0, trace1]

layout = go.Layout(
     title='Unemployment and Suicide Rate from 2010 to 2014',
    titlefont = dict(
         family='Old Standard TT, serif',
        size = 25,
        color = 'Black'
    ),
    
    xaxis=dict(
        title='Year',
        titlefont=dict(
            family='Old Standard TT, serif',
            size=18,
            color='black'
                ),
        
        dtick = 1,
        showticklabels=True,
        
        tickfont=dict(
            family='Old Standard TT, serif',
            size=14,
            color='black'
                ),
            ),

     yaxis=dict(
        title='Rate',
        titlefont=dict(
            family='Old Standard TT, serif',
            size=18,
            color='black'
                ),
        showticklabels=True,
        tickfont=dict(
            family='Old Standard TT, serif',
            size=14,
            color='black'
                    ),
             )
)

fig = go.Figure(data=data, layout=layout)
pyo.iplot(fig)
```

<font size="3">The chart above discloses the relationship between Unemployment and suicide, where it proves that if there were an increase in unemployment rate there will be an increase in the suicide rate and vice versa. Hence, it can be concluded that the suicide rate is proportional to the unemployment rate.</font>

# Suicide Rate By Country and Year 

<font size="3">Here the suicide values were extracted from the merged data based on the country and reshaped the data and cleadned null values to plot suicide rate for each country by years.</font>

```python
# groupping data by country and years
suicountdata = cleandata.groupby(['Country','Year'],as_index=False).agg({"suicide_Rate": "sum"},
                                                                       inplace=True)
```

```python
suicountdata.head()
```

```python
# Change the Year dataFrame from integer to object to be able to plot it on bar chart
suicountdata['Year'] = suicountdata['Year'].astype(str)
```

```python
#Reshape columns to plot using Pandas.DataFraem.pivot
suiyears = suicountdata.pivot( index= 'Country',columns='Year')['suicide_Rate'].reset_index()
suiyears.columns.name = None
```

```python
#Dataset after reshapping
suiyears.head()
```

```python
# Dropping null values
suiyears.dropna(inplace=True)
```

```python
# Plot suicide rate in bar chart by country and years with plotly.graph_objs.Bar

data = [
    go.Bar(
        x=suiyears['Country'],
        y=suiyears['2010'],
        name='Year 2010',
        text = suiyears['Country']
    ),
    go.Bar(
        x=suiyears['Country'], 
        y=suiyears['2011'],
        name='Year 2011',
        text = suiyears['Country']
    ),
    go.Bar(
       x=suiyears['Country'], 
        y=suiyears['2012'],
        name='Year 2012',
        text = suiyears['Country']
    ),
    go.Bar(
        x=suiyears['Country'], 
        y=suiyears['2013'],
        name='Year 2013',
        text = suiyears['Country']
    ),
    go.Bar(
        x=suiyears['Country'], 
        y=suiyears['2014'],
        name='Year 2014',
        text = suiyears['Country']
    )
    

]

layout = go.Layout(
    barmode='stack',
    title='Suicide Rate By Country and Year',
    titlefont = dict(
        size = 25,
        color = 'Black',
            family='Old Standard TT, serif',        
    ),
    xaxis=go.layout.XAxis(
        tickangle= 45,
        color = 'Black',
        title = "Country",
        tickfont=dict(
            family='Old Standard TT, serif',
            size=14,
            color='black'
                    ),
    ),
     yaxis= dict(
         color = 'black',
         title = "Suicide Rate",
         tickfont=dict(
            family='Old Standard TT, serif',
            size=14,
            color='black'
                    ),
    ),
)

fig = go.Figure(data=data, layout=layout)
pyo.iplot(fig)
```

<font size="3"> From the chart above it is clear that <strong>Suriname</strong> has the most suicide rate Compared to other countries from year 2010 to 2013. Whereas, in year 2014 <strong>Latvia</strong> had the most suicide rate. On the other hand, <strong>Bahrain</strong> and <strong>South Africa</strong> has the least number of suicides for all five years. </font>

# Mapping Youth Suicide and Unemployment Rates from 2010 to 2014

```python
#group the data by country and year to sum the suicide and unemployment rates
grouping = cleandata.groupby('Country', as_index=False).agg({"suicide_Rate": "sum",
                                           "Unemployment_Rate": "sum"},inplace=True)
```

```python
grouping.head()
```

```python
# Plot suicide rate in map

data1 = [ dict(
   
        type = 'choropleth',
        locations = grouping['Country'],
        locationmode = 'country names',
        z = grouping['suicide_Rate'],
        text = grouping['Country'],
    
        autocolorscale = True,
        reversescale = False,
        marker = dict(
            line = dict (
                color = 'rgb(190,190,190)',
                width = 0.5
            ) ),
        colorbar = dict(
            autotick = False,
            title = 'Suicide Rate',
        titlefont = dict(
        family='Old Standard TT, serif',
        size = 15,
        color = 'Black'
    ),),
)]


layout =  dict(
    title = 'Youth Suicide Rate by Country<br> from 2010 to 2014',
    titlefont = dict(
        family='Old Standard TT, serif',
        size = 25,
        color = 'Black'
    ),
       
 geo = go.layout.Geo(
        showframe = False,
        showcoastlines = True,
))

fig = dict( data=data1, layout=layout )
pyo.iplot( fig, validate=False)
```

<font size="3"><strong>Suriname, Guyana</strong> and <strong>Lithuania</strong>, from the visualization above, had the most number of suicides from 2010 to 2014 which are denoted in dark red. while <strong>Oman</strong> has the least numbers of suicides in gray.</font>

# Mapping Youth Unemployment Rate from 2010 to 2014

```python
# Plot unemployment rate in map

data = [ dict(
        type = 'choropleth',
        locations = grouping['Country'],
        locationmode = 'country names',
        z = grouping['Unemployment_Rate'],
        text = grouping['Country'],
        colorscale='Viridis',
        autocolorscale = False,
        reversescale = True,
    
        marker = dict(
            line = dict (
                color = 'rgb(180,180,180)',
                width = 0.5
            )),
        colorbar = dict(
            autotick = True,
            title = 'Unemployment Rate',
          titlefont = dict(
        family='Old Standard TT, serif',
        size = 15,
        color = 'Black'
    ),)
    
)]


layout = dict(
        title = 'Youth Unemployment Rate by Country<br> from 2010 to 2014',
        titlefont = dict(  
        family='Old Standard TT, serif',
        size = 25,
        color = 'Black'
    ),
     
 geo = go.layout.Geo(
        showframe = False,
        showcoastlines = True,
),

legend=go.layout.Legend(
                orientation="h",))



fig = dict( data=data, layout=layout )
pyo.iplot( fig, validate=False)
```

<font size="3">The map clearly Identifies the most unemployment rate countries which are <strong>Spian, South Africa</strong> and <strong>Serbia</strong>, pointed with dark blue colore. On the other hand, the least country with least unemployment rate is <strong>Qatar</strong> with a rate of only 14.6 </font>
